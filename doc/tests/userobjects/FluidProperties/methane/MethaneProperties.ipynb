# Thermophysical properties of CH$_4$ #

Calculate the thermophysical properties of methane.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Density ##

Calculate the density of CH$_4$ as a function of pressure and temperature using the ideal gas formulation.

Read in MOOSE data and NIST reference data.

```python
moosedata = pd.read_csv('gold/MethaneProperty_density_vpp_0001.csv')
nistdata = pd.read_csv('gold/nist_t100.csv', sep = '\t', header=None)
```

```python
plt.plot(5*moosedata.x, moosedata.MethanePropertyAuxVariable, label='MOOSE')
plt.plot(nistdata.icol(1), nistdata.icol(2),'o', label='NIST')
plt.legend(loc='upper left');
```

## Viscosity ##

Calculate the density of viscosity as a function of temperature.

Correlations from Irvine Jr, T. F. and Liley, P. E. *Steam and Gas Tables with Computer Equations* (2004).

Read in MOOSE data and NIST reference data.

```python
moosedata = pd.read_csv('gold/MethaneProperty_viscosity_vpp_0001.csv')
nistdata = pd.read_csv('gold/nist_p1.csv', sep = '\t', header=None)
```

```python
plt.plot(100*moosedata.x, moosedata.MethanePropertyAuxVariable, label='MOOSE')
plt.plot(nistdata.icol(0), nistdata.icol(11),'o', label='NIST')
plt.legend(loc='upper left');
```
